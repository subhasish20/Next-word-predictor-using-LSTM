```python
# creating an object of tokenizer class
tokenizer = Tokenizer(oov_token='') # if any word came out side the train data it will simple print nothing
```

```python
# makint the input sequences
```

```python
input_sequences = []
for sentence in text.split('\n'):
  if sentence.strip():
    tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

    for i in range(1,len(tokenized_sentence)):
      input_sequences.append(tokenized_sentence[:i+1])
```

```python
# for 5 the next word will be 81
# for 5 and 81 the next word will be 21
# for 5,81and 21 the next word will be 219

"""
so if X = 5 then y = 81
if X = 5,81 then y = 21
if X = 51,81,21 then y = 219
"""
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense,Dropout
```

```python
vocab = len(tokenizer.word_index.keys()) +1
max_length = max([len(x) for x in input_sequences])
```

```python
print(f"the vocab size of the number of words in the data is {vocab}")
print(f"the maximum lenght of sentences in the data is {max_length}")
```

```python

"""
FOR Embedding layer:
    input_dim:
    - Meaning: This is the size of the vocabulary, i.e., the number of unique words or tokens in your dataset

    output_dim:
    - Meaning: This specifies the dimensionality of the output embedding vectors for each word.

    input_length:
    - Meaning: This specifies the length of the input sequences to the Embedding layer (i.e., the max number of words per sequence).
"""
```

```python
model = Sequential()
```

```python
# embedding layer
model.add(Embedding(input_dim=2480, output_dim=100, input_shape=(50,))) # or input_dim=vocab,output_dime=100,input_shape=(maxlength-1,)

# first LSTM layer with Dropout
model.add(LSTM(160, return_sequences=True, activation="tanh"))
model.add(Dropout(0.2))

# second LSTM layer with Dropout
model.add(LSTM(300, return_sequences=True, activation="tanh"))
model.add(Dropout(0.2))

# third LSTM layer
model.add(LSTM(300, activation="tanh"))
model.add(Dropout(0.2))

# output Dense layer
model.add(Dense(2480, activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X, y,epochs=100)
```

```python
import time
text = "I just wanted"

for i in range(10):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=51, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```
