```python
import tensorflow as tf
from tensorflow import keras
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([text])
```

```python
tokenizer.word_index
```

```python
# converting into numbers
```

```python
for sentence in text.split('\n'):
  if sentence.strip():
    print(tokenizer.texts_to_sequences([sentence])[0])
```

```python
# makint the input sequences
```

```python
input_sequences = []
for sentence in text.split('\n'):
  if sentence.strip():
    tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

    for i in range(1,len(tokenized_sentence)):
      input_sequences.append(tokenized_sentence[:i+1])
```

```python
# for 5 the next word will be 81
# for 5 and 81 the next word will be 21
# for 5,81and 21 the next word will be 219

"""
so if X = 5 then y = 81
if X = 5,81 then y = 21
if X = 51,81,21 then y = 219
"""
```

```python
input_sequences
```

```python
# finding the maximun length sequence
```

```python
max_len = max([len(x) for x in input_sequences])
max_len
```

```python
# applying padding
# padding by adding zero before all the list for making same shape
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
padded_input_sequences
```

```python
padded_input_sequences.shape
```

```python
# getting x and y for feature and target column
```

```python
X = padded_input_sequences[:,:-1] # it will not take the last value (Which is the target (or X))
X
```

```python
X.shape
```

```python
y = padded_input_sequences[:,-1] # taking the last value as a output(or y)
```

```python
y
```

```python
y.shape
```

```python
len(tokenizer.word_index)
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y, num_classes=13216)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
```

```python
"""
model.add(Embedding(13216, output_dim=200,mask_zero=True, input_shape=(42,)))
model.add(LSTM(128,return_sequences=True,activation='relu'))
model.add(LSTM(256,return_sequences=True))
model.add(LSTM(128))
model.add(Dense(13216, activation='softmax'))"""
```

```python
model.add(Embedding(13216, output_dim=200, input_shape=(42,)))
model.add(LSTM(128, return_sequences=True, activation="tanh"))
model.add(LSTM(256, return_sequences=True, activation="tanh"))
model.add(LSTM(128, activation="tanh"))
model.add(Dense(13216, activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X,y,epochs=70)
```

```python
import time
text = "this"

for i in range(20):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=56, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```
